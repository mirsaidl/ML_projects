```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
size = np.array([1.0,2.0,1.2])
price = np.array([300,500,340])


print(size)
print(price)
```

```python
plt.scatter(size,price, marker='x', c='red')
plt.title('Housing Prices')
plt.ylabel('Price(in 1000s of dollars)')
plt.xlabel('Size(1000 sqft)')

plt.show()
```

```python
k = 100
b = 100

print(f'k: {k}')
print(f'b: {b}')
```

```python
def compute(x , k, b):
    m = x.shape[0]
    y = np.zeros(m) # 
    for i in range(m):
        y[i] = k * x[i] + b
    
    return y
```

```python
# Creating a 1-dimensional array of size 3 filled with zeros
array = np.zeros(3)
print(array) # zeros will fill array with zeros
```

```python
func = compute(size, k, b)

plt.plot(size,func, 'b', label = 'Our Prediction')
plt.scatter(size,price, marker='x', c='red', label='Actual Values')

plt.title('Housing Prices')
plt.ylabel('Price(in 1000s of dollars)')
plt.xlabel('Size(1000 sqft)')


plt.legend()
plt.show()
```

```python
k = 200
b = 100
# actual values

func = compute(size, k, b)

plt.plot(size,func, 'b.-', label = 'Our Prediction')
plt.scatter(size,price, marker='x', c='red', label='Actual Values')

plt.title('Housing Prices')
plt.ylabel('Price(in 1000s of dollars)')
plt.xlabel('Size(1000 sqft)')


plt.legend()
plt.show()
```
